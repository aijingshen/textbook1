# Oxygen transfer through artificial skin (Dermagraft)

```python
# Import needed packages
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from ipywidgets import interact, FloatSlider
```

## In class we derived an equation for the concentration of oxygen in the skin dermagraft. 
The dimensionless equation for the concentration profile \( C(x) \) is:

$\frac{C(x)}{C_0} = 1 - Thiele \left( 1 - \left( \frac{x}{L} \right)^2 \right)$

where:
- $C(x)$ is the concentration at position $x$
- $C_0$ is the concentration at $x = L$
- $L$ is the thickness of the dermagraft
- $Thiele (\frac{R L^2}{2D C_0})$ is a dimensionless parameter that compares the reaction rate to diffusion
- $R$ is the oxygen consumption rate in this case

This equation describes the steady-state concentration profile in a 1D domain with a constant reaction rate and diffusion.

## Plot oxygen concentration as a function of distance

```python
# Define the dimensionless function C(xi)/C0 = 1 - Thiele * (1 - xi^2)
def dimensionless_concentration(xi, Thiele):
    return 1 - Thiele * (1 - xi**2)

# Define a function to plot U(xi) vs xi
def plot_dimensionless_solution(Thiele):
    # Create an array of xi values between 0 and 1
    xi = np.linspace(-1, 1, 100)
    
    # Calculate the corresponding U(xi) values
    C_xi = dimensionless_concentration(xi, Thiele)
    
    # Plot the result
    plt.figure(figsize=(8, 6))
    plt.plot(xi, C_xi, label=f'Thiele = {Thiele}', color='blue')
    plt.xlabel('$x/L$', fontsize=14)
    plt.ylabel('$C(x)/C_0$', fontsize=14)
    plt.ylim(0,1)
    plt.title('Dimensionless Concentration Profile', fontsize=16)
    plt.grid(True)
    plt.legend()
    plt.show()
  
# Example plot with beta = 0.5
plot_dimensionless_solution(0.5)
```

## The Thiele dimensionless constant 


Thiele = $(\frac{R L^2}{2D C_0})$ that is useful for understanding the relative contributions of the reaction and diffusion

Let's plot the concentration of oxygen in the dermagraft for small, medium, and large Thiele

```python
# Define a function to plot C(xi)/C0 vs xi for different Thiele values
def plot_dimensionless_solutions():
    # Define Thiele modulus values for high, middle, and small
    Thiele_values = [0.1, 0.5, 2]
    labels = ['Small Thiele', 'Middle Thiele', 'High Thiele']
    
    # Create an array of xi values between -1 and 1
    xi = np.linspace(-1, 1, 100)
    
    # Plot each Thiele case
    plt.figure(figsize=(8, 6))
    for Thiele, label in zip(Thiele_values, labels):
        C_xi = dimensionless_concentration(xi, Thiele)
        plt.plot(xi, C_xi, label=f'{label} (Thiele = {Thiele})')
    
    # Plot settings
    plt.xlabel('$x/L$', fontsize=14)
    plt.ylabel('$C(x)/C_0$', fontsize=14)
    plt.ylim(0, 1)
    plt.title('Dimensionless Concentration Profile', fontsize=16)
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function to plot the solutions
plot_dimensionless_solutions()

```

Look at the large Thiele case. In this case, ${R L^2} > {2D C_0}$. 

The concentration of [O2] drops quickly from the interface with the media and reaches 0 at x/L ~0.75. 

This indicates that a large part of the Dermagraft is without oxygen. 

As a result, cells will die, and the Dermagraft will not promote healing.
