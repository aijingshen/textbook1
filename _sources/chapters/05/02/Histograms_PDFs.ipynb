# Example 1: Rolling 2 dice
Let's roll two dice and add the values together. 

```python
# Function to roll two dice and return their sum
def roll_two_dice():
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1 + die2
```

## We roll once and get a number

```python
# Roll the dice 10 times and store the sums
dice_sums = [roll_two_dice() for _ in range(1)]
plt.hist(dice_sums, bins=range(0, 14), edgecolor='black', align='left')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Frequency')
plt.title('Histogram of Dice Roll Sums')
plt.xticks(range(0, 15))  # Setting x-ticks to show values up to 14 for better spacing
plt.show()
```

## We decide to roll 10 times

```python
dice_sums = [roll_two_dice() for _ in range(10)]
# Plotting the histogram with bins from 0 to 13 to include 12
plt.hist(dice_sums, bins=range(0, 14), edgecolor='black', align='left')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Frequency')
plt.title('Histogram of Dice Roll Sums')
plt.xticks(range(0, 15))  # Setting x-ticks to show values up to 14 for better spacing
plt.show()
```

## In this case we actually know the underlying probability of each sum


These probabilities can be calculated theoretically by counting the number of ways each possible sum can occur when rolling two six-sided dice. Since each die has 6 faces, there are  6 $\times$ 6 = 36  total possible outcomes. 

**Total Outcomes:**  
There are 36 total possible outcomes when rolling two six-sided dice.

**Number of Ways to Achieve Each Sum:**
   - Sum of 2:  (1,1)  — 1 way
   - Sum of 3:  (1,2), (2,1)  — 2 ways
   - Sum of 4:  (1,3), (2,2), (3,1)  — 3 ways
   - Sum of 5:  (1,4), (2,3), (3,2), (4,1)  — 4 ways
   - Sum of 6:  (1,5), (2,4), (3,3), (4,2), (5,1)  — 5 ways
   - Sum of 7:  (1,6), (2,5), (3,4), (4,3), (5,2), (6,1)  — 6 ways
   - Sum of 8:  (2,6), (3,5), (4,4), (5,3), (6,2)  — 5 ways
   - Sum of 9:  (3,6), (4,5), (5,4), (6,3)  — 4 ways
   - Sum of 10:  (4,6), (5,5), (6,4)  — 3 ways
   - Sum of 11:  (5,6), (6,5)  — 2 ways
   - Sum of 12:  (6,6)  — 1 way

**Calculate Probabilities:**
The probability of each sum is the number of ways to achieve that sum divided by the total number of outcomes (36).

Let’s compute these probabilities:

Here are the theoretical probabilities for each possible sum when rolling two six-sided dice, calculated based on the number of ways each sum can be achieved:

- Sum of 2:  1/36 $\approx$ 0.028  (2.78%)
- Sum of 3:  2/36 $\approx$ 0.056  (5.56%)
- Sum of 4:  3/36 $\approx$ 0.083  (8.33%)
- Sum of 5:  4/36 $\approx$ 0.111  (11.11%)
- Sum of 6:  5/36 $\approx$ 0.139  (13.89%)
- Sum of 7:  6/36 $\approx$ 0.167  (16.67%)
- Sum of 8:  5/36 $\approx$ 0.139  (13.89%)
- Sum of 9:  4/36 $\approx$ 0.111  (11.11%)
- Sum of 10: 4/36 $\approx$ 0.083  (8.33%)
- Sum of 11: 3/36 $\approx$ 0.056  (5.56%)
- Sum of 12: 1/36 $\approx$ 0.028  (2.78%)

## How does the measured probability of this event compare to the theoretical? 

```python
# Count the frequency of each sum
sum_counts = {i: dice_sums.count(i) for i in range(2, 13)}

# Convert the counts into probabilities by dividing by the total number of rolls
total_rolls = len(dice_sums)
sum_probabilities = {sum_value: count / total_rolls for sum_value, count in sum_counts.items()}

# Theoretical probabilities for each sum
theoretical_probabilities = {
    2: 1/36,
    3: 2/36,
    4: 3/36,
    5: 4/36,
    6: 5/36,
    7: 6/36,
    8: 5/36,
    9: 4/36,
    10: 3/36,
    11: 2/36,
    12: 1/36
}

# Plotting the empirical probabilities as a bar chart
plt.bar(sum_probabilities.keys(), sum_probabilities.values(), edgecolor='black', label='Empirical')

# Plotting the theoretical probabilities as a line plot
plt.plot(theoretical_probabilities.keys(), theoretical_probabilities.values(), color='red', marker='o', linestyle='-', label='Theoretical')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('Probability Distribution of Dice Roll Sums')
plt.xticks(range(2, 13))  # Setting x-ticks to show possible sums
plt.legend()
plt.show()
```

## What if we count a lot of dice rolls? 

```python
rolls = 100000
dice_sums = [roll_two_dice() for _ in range(rolls)]

# Count the frequency of each sum
sum_counts = {i: dice_sums.count(i) for i in range(2, 13)}

# Convert the counts into probabilities by dividing by the total number of rolls
total_rolls = len(dice_sums)
sum_probabilities = {sum_value: count / total_rolls for sum_value, count in sum_counts.items()}

# Plotting the empirical probabilities as a bar chart
plt.bar(sum_probabilities.keys(), sum_probabilities.values(), edgecolor='black', label='Empirical')

# Plotting the theoretical probabilities as a line plot
plt.plot(theoretical_probabilities.keys(), theoretical_probabilities.values(), color='red', marker='o', linestyle='-', label='Theoretical')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('Probability Distribution of Dice Roll Sums')
plt.xticks(range(2, 13))  # Setting x-ticks to show possible sums
plt.legend()
plt.show()
```

So, we see that the empirical probabilities EXACTLY matches the measured probabilities when we take more and more measurements
