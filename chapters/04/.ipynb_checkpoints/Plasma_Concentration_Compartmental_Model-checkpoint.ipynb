
# Single Compartment Model Practice Problem

This notebook provides a detailed breakdown of the formula and step-by-step solution for a pharmacokinetic problem involving a one-compartment model.

Question 1: Assuming a one compartment linear pharmacokinetic model, with k = 0.3 1/hr, ka = 0.9 1/ hr, F = 70%; and V = 20 L, calculate the plasma concentration (in mg/L) at 1 hours after a 1000 mg oral dose?


The formula for plasma concentration \( C(t) \) of a drug, administered orally and assuming a one-compartment model with first-order absorption and elimination, is given by:
$$
C(t) = \frac{F \cdot \text{Dose}}{V} \cdot \frac{ka}{ka - k} \cdot (e^{-kt} - e^{-kat})
$$


| Parameter                  | Value          |
|----------------------------|----------------|
| Dose (Dose)                | 1000 mg        |
| Bioavailability (F)        | 0.70           |
| Elimination rate constant (k) | 0.3 1/hr    |
| Absorption rate constant (ka) | 0.9 1/hr    |
| Volume of distribution (V) | 20 L           |
| Time (t)                   | 1 hour         |
| Plasma Concentration (C(1))| 52.326 mg/L    |


Substitute these into the formula and calculate at \( t = 1 \) hour:

1. **Plasma Concentration Factor**:
   $$
   C_{factor} = \frac{0.7 \cdot 1000 \, \text{mg}}{20 \, \text{L}} = 35 \, \text{mg/L}
   $$

2. **Absorption-Elimination Ratio**:
   $$
   \frac{ka}{ka - k} = \frac{0.9}{0.9 - 0.3} = 1.5
   $$

3. **Exponential Terms for Elimination and Absorption**:
   $$
   e^{-k \cdot 1} = e^{-0.3} \approx 0.7408
   $$
   $$
   e^{-ka \cdot 1} = e^{-0.9} \approx 0.4066
   $$

4. **Final Plasma Concentration at \( t = 1 \) hour**:
   $$
   C(1) = 35 \cdot 1.5 \cdot (0.7408 - 0.4066)
   $$
   $$
   C(1) = 35 \cdot 1.5 \cdot 0.3342
   $$
   $$
   C(1) = 17.5025 \, \text{mg/L}
   $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the urea concentration over time
def plasma_concentration(t, F, Dose, V, k_a, k):
    k_total = k_a - k  # combined elimination rate constant
    steady_state_concentration = F*Dose*k_a / (V * k_total)
    return steady_state_concentration * (np.exp(-k * t)-np.exp(-k_a*t))

# Parameters
F = 0.7  
V = 20  # Volume of distribution (L)
k = 0.3  # Elimination rate constant (1/hr)
k_a = 0.9  # Absorption rate constant (1/hr)
Dose = 1000 

# Time points from 0 to 24 hours
t = np.linspace(0, 24, 500)

# Calculate the urea concentration over time
C_values = plasma_concentration(t, F, Dose, V, k_a, k)

# Plot the graph
plt.plot(t, C_values, label=r'$C(t) = \frac{F \cdot \text{Dose}}{V} \cdot \frac{ka}{ka - k} \cdot (e^{-kt} - e^{-kat})$')
plt.axvline(x=1, color='red', linestyle='--', label='t = 1 hour')

# Plot formatting
plt.xlabel('Time (hours)')
plt.ylabel('Plasma Concentration (mg/L)')
plt.title('Plasma Concentration Over Time During Dialysis')
plt.legend()
plt.ylim(0, None)
plt.grid(True)
plt.show()
```
