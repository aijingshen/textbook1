A 61 kg patient needs enough morphine to relieve their post-surgical pain during sleep (8 hours). How much morphine should be given to the patient before bed so that the concentration of morphine in their system doesn't drop below the pain-relieving concentration $8.2 \times 10^{-3} \mathrm{mg} / \mathrm{L}$?

The volume of distribution for morphine $V_{d, \text { morphine }}$ is approximately 1.0 $\mathrm{kg} / \mathrm{L}$, or 61 L (for this 61 kg patient).

This patient's clearance rate for morphine $k=0.351 / \mathrm{hr}$. Below is the one-compartment model describing this system.



## Central compartment

$V_{d}=61 \mathrm{~L}$ Initial dose $\mathrm{C}_{0}=$ ?

The rate of change of morphine concentration in the central compartment would be:
$$
\begin{aligned}
\frac{d c}{d t} & =-k c \\
\frac{d c}{c} & =-k d t
\end{aligned}
$$

Integrate both sides to get:

$$
\begin{aligned}
\int \frac{d c}{c} & =-k \int d t \\
\ln c & =-k t+A
\end{aligned}
$$

where A is an integration constant. This is equivalent to:

$$
\ln c=-k t+\ln A
$$

Why can we switch out A for $\ln A$? Why did we do that?

$$
\begin{gathered}
\ln c-\ln A=-k t \\
\ln \frac{c}{A}=-k t \\
c(t)=A e^{-k t}
\end{gathered}
$$


# Required Initial Dose
<div class="alert alert-info"> 
QUESTION: How do we calculate the required initial dose of morphine?
</div>

We know two things: 
- At $t=0, c(t)$ must be $c_{0}$ (the initial concentration)
- At $t=8$ hours, the concentration can be no less than $8.2 \circ 10^{-3} \mathrm{mg} / \mathrm{L}$.

We use the first to solve for A .
$$
c_{0}=A e^{0}
$$

Which gives us $\mathrm{A}=c_{0}$ and,
$$
c(t)=c_{0} e^{-k t}
$$

Now, we solve for a $C_{0}$ that will give us exactly the effective pain-relieving concentration after 8 hours. At this initial dose, the concentration of morphine will just start to become ineffective as the patient wakes up.

$$
8.2 \circ 10^{-3} \mathrm{mg} / \mathrm{L}=c_{0} e^{-(0.35 / h r)(8.0 h r)}
$$

Solving for $c_{0}$ gives us a concentration of $0.134846 \ldots \mathrm{mg} / \mathrm{L}$ (accounting for sig figs in the last step). 

To achieve this concentration in the body, one must inject $(0.134846 \ldots \mathrm{mg} / \mathrm{L})(61 \mathrm{~L})$, or 8.2 mg. In reality, since we've made several simplifying assumptions in our compartmental model, I would interpret this as "somewhere around 8 mg."

```python
<div class="alert alert-info"> 
QUESTION: How do we calculate the required initial dose of morphine?
</div>

QUESTION: Before giving a patient this dose, what else should we consider?

The typical dose for morphine is $10-30 \mathrm{mg}$. Overdose occurs between 60 mg (hypersensitive) and 200 mg (typical), though it can be much higher for those who have been using morphine regularly.

QUESTION: An 80 kg patient comes in with the same problem - you're asked to calculate a minimum effective dose. What do you need to do differently?

For an 80 kg patient, the volume of distribution for morphine $V_{d, \text { morphine }}$ will be 80 L .

If the clearance rate is proportional to the surface area of the patient, then,
$$
\frac{k_{80 \mathrm{~kg}}}{k_{61 \mathrm{~kg}}}=\frac{1.73\left(\frac{80 \mathrm{~kg}}{70}\right)^{0.73}}{1.73\left(\frac{61 \mathrm{~kg}}{70}\right)^{0.73}}=\left(\frac{80}{61}\right)^{0.73}=1.2
$$

The clearance rate for the 80 kg patient will be $\mathrm{k}=0.421 / \mathrm{hr}$.

\section*{Chapter 2}

\section*{Compartmental model of iodine transport to the thyroid}

The thyroid can take up and store iodine. A simple compartmental model of what happens when radioactive iodine is accidentally injested is shown below.

![](https://cdn.mathpix.com/cropped/2024_09_09_d18d4cecd0b381a9bd9dg-6.jpg?height=312&width=1202&top_left_y=1568&top_left_x=451)

QUESTION: What processes are left out of the compartmental model? Why would I leave them out?

Our compartmental model assumes the following:
- Once ingested, iodine moves quickly through the digestive system into the central compartment
- The rate of iodine release from the thyroid is slow compared to the rates
of excretion and transport to the thyroid
- The volume of distribution is approximately constant.
- The thyroid doesn't absorb so much iodine that it becomes "full"
- Before the ingestion, the thyroid was free of radioactive iodine.

So, keep in mind that our model is not good for "short times" (on the order of minutes) because it assumes that movement from the mouth to the stomach, and from the stomach to the central compartment, are instantaneous. This probably takes a few minutes. The model is also not good for "long times" (on the order of weeks) during which we'd need to take into account iodine release from the thyroid. Our model is probably good for "medium times", on the order of hours or maybe a few days - which is fine! That's what we plan on using it for.

I'm a graduate student working with radioactive Iodine-125. To determine if I've been accidentally exposed, after working with the material, I wait an adequate amount of time for the Iodine-125 to end up in my thyroid, and then have my thyroid scanned for radioactivity.

QUESTION: How long after working with Iodine-125 should I wait to have my thyroid scanned?

The rate of change of Iodine-125 in the central compartment would be:
$$
\frac{d c}{d t}=-\left(k_{t}+k_{e}\right) c
$$

Solve as before to get $\mathrm{c}(\mathrm{t})$, the concentration of iodine in the central compartment:
$$
c(t)=c_{0} e^{-\left(k_{t}+k_{e}\right) t}
$$

QUESTION: How do I get from concentration in the central compartment
to amount in the thyroid?
If I is the amount of iodine stored in the thyroid, then:
$$
\frac{d I}{d t}=k_{t} c V_{d}
$$

QUESTION: What is the $V_{d}$ doing in the equation?
$$
\underbrace{\frac{d I}{d t}}_{\frac{\text { amount }}{\text { time }}}=\underbrace{k_{t}}_{\frac{1}{\text { time }}} \underbrace{c}_{\frac{\text { amount }}{\text { volume }}} \underbrace{V_{d}}_{\text {volume }}
$$

Note that we have an equation for $\mathrm{c}(\mathrm{t})$. Plugging it into the equation gives,
$$
\frac{d I}{d t}=k_{t} V_{d} c_{0} e^{-\left(k_{t}+k_{e}\right) t}
$$

Integrating and solving for $\mathrm{I}(\mathrm{t})$ gives,
$$
I(t)=-\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}} e^{-\left(k_{t}+k_{e}\right) t}+A
$$

QUESTION: How do we solve for A?
Given that at $t=0, I(t)=0$,
$$
0=-\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}}+A
$$

Which gives,
$$
I(t)=\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}}\left(1-e^{-\left(k_{t}+k_{e}\right) t}\right)
$$

Let's say that I receive an initial dosage of 2.0 mCi of Iodine-125. The initial concentration in my central compartment would be $C_{0}=(2.0 \mathrm{mCi}) /(20 \mathrm{~L})$. The amount of Iodine-125 in my thyroid would be,


QUESTION: How long after working with radioactive Iodine-125 should I wait before getting my thyroid checked?
```
