## Single Compartment Models

Pharmacokinetic models often divide the body into compartments where drugs are distributed.  

```{admonition} One compartment model
The simplest model is the **one-compartment model**, where the drug is assumed to be distributed instantaneously throughout a single, homogenous compartment (the body).

![one-compartment](../../../images/compartments1.png)

$$
c(t)=c_0 e^{-k_e t}
$$

```


In a single compartment model rate of change of drug in the central compartment is:

$$
\begin{aligned}
\frac{d c}{d t} & =-k c \\
\frac{d c}{c} & =-k d t
\end{aligned}
$$

Integrate both sides to get:

$$
\begin{aligned}
\int \frac{d c}{c} & =-k \int d t \\
\ln c & =-k t+A*
\end{aligned}
$$

We can switch out A* for $\ln A$, making solving easier. Both are constants. 

$$
\ln c=-k t+\ln A
$$

$$
\begin{gathered}
\ln c-\ln A=-k t \\
\ln \frac{c}{A}=-k t \\
c(t)=A e^{-k t}
\end{gathered}
$$

At time $t=0, c(t)$ must be $c_{0}$ (the initial concentration). We can now solve for A.

$$
c_{0}=A e^{0}
$$

Which gives us $\mathrm{A}=c_{0}$ and

$$
c(t)=c_{0} e^{-k t}
$$

This is an exponential decay model. 

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for concentration over time
def concentration(t, C0, k):
    return C0 * np.exp(-k * t)

# Parameters
C0 = 1.0  # Initial concentration (mg/L)
k = 0.2  # Elimination rate constant (1/hr)

# Time points from 0 to 24 hours
t = np.linspace(0, 24, 500)

# Calculate the concentration over time
C_values = concentration(t, C0, k)

# Plot the graph
plt.plot(t, C_values, label=r'$C(t) = C_0 e^{-kt}$')

# Add labels and title
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (mg/L)')
plt.title('Exponential Decay of Drug Concentration Over Time')

# Add a grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
````
