# Normal distribution
The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about its mean. It's characterized by its bell-shaped curve, which is defined by two parameters: the mean (μ) and the standard deviation (σ). 

### Mathematical Formula:
The probability density function (pdf) of the normal distribution is given by:

$$
N(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Where:
-  $x$  is a variable,
-  $\mu$  is the mean,
-  $\sigma$  is the standard deviation,
-  $e$  is the base of the natural logarithm,
-  $\pi$  is a constant (approximately 3.14159).

### Key Features:
- **Mean (μ):** The peak of the curve is at the mean, which is also the center of the distribution. The mean represents the average value of the data set.
- **Standard Deviation (σ):** This measures the spread or dispersion of the data. A smaller standard deviation results in a narrower, taller curve, while a larger standard deviation produces a wider, flatter curve.
- **Total Area:** The total area under the curve is equal to 1, which represents the entirety of the probability.


```python
# Define the function for the normal distribution
def normal_distribution(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Create an array of x values
x_values = np.linspace(-10, 10, 400)

# Define different sets of parameters (mu, sigma)
parameter_sets = [
    (0, 1),   # mu = 0, sigma = 1
    (0, 2),   # mu = 0, sigma = 2
    (-2, 1),  # mu = -2, sigma = 1
    (2, 3)    # mu = 2, sigma = 3
]

# Set up colors for each plot
colors = ['blue', 'green', 'red', 'purple']

# Plot the functions for different parameters
plt.figure(figsize=(10, 6))
for (mu, sigma), color in zip(parameter_sets, colors):
    y_values = normal_distribution(x_values, mu, sigma)
    plt.plot(x_values, y_values, color=color, label=f'$\mu={mu}, \sigma={sigma}$')

# Add title and labels
plt.title('Normal Distribution (PDF) for Different $\mu$ and $\sigma$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Probability density functions 

```python
from scipy.stats import norm

# Define the function for the normal distribution
def normal_distribution(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Parameters for the normal distribution
mu = 0
sigma = 1

# Create an array of x values
x_values = np.linspace(-5, 5, 400)

# Calculate the y values using the normal distribution function
y_values = normal_distribution(x_values, mu, sigma)

p_low = 0.2
p_high = 0.7

# Define the range to shade
x_shade = np.linspace(p_low, p_high, 100)
y_shade = normal_distribution(x_shade, mu, sigma)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$', color='blue')

# Shade the area between 0.5 and 0.75
plt.fill_between(x_shade, y_shade, color='orange', alpha=0.5, label='Area')

# Add labels and title
plt.title('Normal Distribution (PDF) with $\mu = 0$ and $\sigma = 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the probability for the area between 0.5 and 0.75
probability = norm.cdf(p_high, mu, sigma) - norm.cdf(p_low, mu, sigma)
print(f"The probability between 0.5 and 0.75 is approximately {probability:.4f}")
```
