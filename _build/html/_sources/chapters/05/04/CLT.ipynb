```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display
path_data = '../../../assets/data/'
```

```python
geom_data = pd.read_excel(path_data + 'geometric_distribution_data.xlsx')

# Plotting the data
plt.hist(geom_data['Event_Count'], bins=range(1, geom_data['Event_Count'].max() + 1), edgecolor='black', density=True)
plt.xlabel('Event Count')
plt.ylabel('Probability')
plt.title('Not normal Distribution Data')
plt.show()
```

## The sample mean
Let's say we take a sample of size n=5, and we take the mean. 

```python
sample = geom_data['Event_Count'].sample(5)
sample.mean()
```

Let's repeat this process ***10 times*** (sample size n=5 each time), get many different means, and plot these means. 

```python
# Define the sample size and number of repetitions
sample_size = 5
num_repeats = 10

# Set up the plot
fig, ax = plt.subplots(figsize=(8, 4))

# Perform sampling with the fixed sample size and the given number of repetitions
sample_means = []
for _ in range(num_repeats):
    sample = geom_data['Event_Count'].sample(sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

# Plotting the distribution of sample means
ax.hist(sample_means, bins=20, edgecolor='black', alpha=0.7, density=True)
ax.set_title(f'Repeats = {num_repeats}, Sample Size = {sample_size}', rotation=0, ha='center')
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Density')
ax.autoscale(enable=True, axis='y')
fig.suptitle('Distribution of Sample Means (Geometric Data)', y=1.05)
plt.tight_layout()
plt.show()
```

## As n increases, the distribution starts to look more and more like the normal distribution

```python
# What happens when we take a lot of sample means? 
sample_size = 5
num_repeats = [10, 100, 1000, 10000, 100000]

# Set up the plot with subplots in one row
fig, axes = plt.subplots(1, len(num_repeats), figsize=(24, 5), sharex=True)

# Perform sampling with the fixed sample size and varying repetitions
for j, repeat in enumerate(num_repeats):
    sample_means = []
    for _ in range(repeat):
        sample = geom_data['Event_Count'].sample(sample_size)
        sample_mean = sample.mean()
        sample_means.append(sample_mean)
    
    # Plotting the distribution of sample means
    ax = axes[j]
    ax.hist(sample_means, bins=20, edgecolor='black', alpha=0.7, density=True)
    ax.set_title(f'Repeats = {repeat}', rotation=0, ha='center')
    ax.set_xlabel('Sample Mean')
    ax.set_ylabel('Density')
    
    # Autoscale y-axis for each subplot
    ax.autoscale(enable=True, axis='y')

# Set the overall title for the figure
fig.suptitle(f'Distribution of Sample Means (Sample Size = {sample_size}) with Varying Repeats', y=1.05)
plt.tight_layout()
plt.show()
```

## As we sample the mean more and more, the distribution of the sample mean has smaller and smaller variance

```python
# Define the sample sizes and fixed number of repetitions
sample_sizes = [10, 100, 500, 1000]
num_repeats = 1000

# Set up the plot with subplots in a single row
fig, axes = plt.subplots(1, len(sample_sizes), figsize=(24, 6), sharex=True)

# Perform sampling with varying sample sizes and a fixed number of repetitions
for i, size in enumerate(sample_sizes):
    sample_means = []
    for _ in range(num_repeats):
        sample = geom_data['Event_Count'].sample(size)
        sample_mean = sample.mean()
        sample_means.append(sample_mean)
    
    # Plotting the distribution of sample means
    ax = axes[i]
    ax.hist(sample_means, bins=20, edgecolor='black', alpha=0.7, density=True)
    ax.set_title(f'Sample Size = {size}', rotation=0, ha='center')
    ax.set_xlabel('Sample Mean')
    ax.set_ylabel('Density')
    
    # Set the x-axis limits to focus around the mean
    ax.set_xlim(0, 15)
    ax.autoscale(enable=True, axis='y')  # Autoscale the y-axis for each graph

# Set the overall title for the figure
fig.suptitle('Distribution of Sample Means with Varying Sample Sizes (Repetitions = 1000)', y=1.02)
plt.tight_layout()
plt.show()
```

The Central Limit Theorem tells us that as $n  \rightarrow large$ the $\mu_{\bar{X}}  \rightarrow \mu$ and as the number of means taken is increased, the variance decreases. This reduces to a remarkable result: the the sampling distribution of the sample mean $ \bar{X} $ approaches a normal distribution as the sample size $ n $ becomes large:

$$
   \bar{X} \sim N\left( \mu, \frac{\sigma^2}{n} \right)
$$

**Mean of the Sampling Distribution:**

$$
   \mu_{\bar{X}} = \mu
$$

**Standard Deviation of the Sampling Distribution (Standard Error):**

$$
   \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}
$$

**Variance of the Sampling Distribution:**

$$
   \sigma_{\bar{X}}^2 = \frac{\sigma^2}{n}
$$
